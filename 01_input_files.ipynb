```python
%matplotlib inline
import matplotlib.pylab as plt
import numpy as np
```

```python
def getdata(ifile, x, y, z, time):
    size  = time*z*y*x
    shape = (time,z,y,x)

    data = np.fromfile(file=ifile, dtype = 'float32', count=size)

    data = data.reshape(shape)
 
    data = data.byteswap()
    return data
```
